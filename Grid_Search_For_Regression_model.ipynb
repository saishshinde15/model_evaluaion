```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('C0-pyrolysis dataset.csv')
```

```python
dataset
```

```python
dataset=dataset.drop(['Combination'],axis=1)
```

```python
dataset
```

```python
dataset = dataset.drop(dataset.tail(2).index)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
sns.heatmap(dataset.corr(),annot=True)
```

```python
dataset
```

```python
X=dataset.drop(['activation energy','bamboo','Heating rate'],axis=1).values
```

```python
X
```

```python
y=dataset['activation energy'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
len(X_train)
```

```python
len(y_train)
```

# Model 1(Drop)

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_lr=lr.predict(X_test)
```

```python
print(np.concatenate((y_lr.reshape(len(y_lr),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,y_lr)
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_lr)
```

```python
from sklearn.metrics import r2_score
R2=r2_score(y_test,y_lr)
```

```python
model_result=(['Logistic_Regression','MAE',MAE,'MSE',MSE,'R2',R2])
```

```python
model_result
```

# Model 2 (1st Best)

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'max_depth': np.arange(1, 20)}

grid_search = GridSearchCV(dt, parameters, cv=5)
grid_search.fit(X_train, y_train)

best_max_depth = grid_search.best_params_['max_depth']
best_dt = DecisionTreeRegressor(max_depth=best_max_depth)
best_dt.fit(X_train, y_train)

y_pred_dt = best_dt.predict(X_test)

MAE_1 = mean_absolute_error(y_test, y_pred_dt)
MSE_1 = mean_squared_error(y_test, y_pred_dt)
R2_1= r2_score(y_test, y_pred_dt)

model_result = (['Logistic_Regression','MAE',MAE,'MSE',MSE,'R2',R2],
                 ['Decision_Tree', 'MAE', MAE_1, 'MSE', MSE_1, 'R2', R2_1])
```

```python
import pandas as pd
model_results = [
    ['Logistic Regression', MAE, MSE, R2],
    ['Decision Tree', MAE_1, MSE_1, R2_1]
]

df = pd.DataFrame(model_results, columns=['Model', 'MAE', 'MSE', 'R2'])
print(df)
```

```python
print(np.concatenate((y_pred_dt.reshape(len(y_pred_dt),1),y_test.reshape(len(y_test),1)),1))
```

```python
y_pred_dt_train = best_dt.predict(X_train) ## training
```

```python
r2_score(y_train, y_pred_dt_train)
```

# Model 3 (3rd Best Accuracy)

```python
from sklearn.ensemble import RandomForestRegressor
rft=RandomForestRegressor()
```

```python
parameters = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10, 15, 100],
    'min_samples_leaf': [1, 2, 5, 10]
}

grid_search = GridSearchCV(rft, parameters, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_rft = RandomForestRegressor(**best_params)
best_rft.fit(X_train, y_train)

y_pred_rft = best_rft.predict(X_test)
```

```python
MAE_3 = mean_absolute_error(y_test, y_pred_rft)
MSE_3 = mean_squared_error(y_test, y_pred_rft)
R2_3 = r2_score(y_test, y_pred_rft)

model_results = [
    ['Logistic Regression', MAE, MSE, R2],
    ['Decision Tree', MAE_1, MSE_1, R2_1],
    ['Random Forest', MAE_3, MSE_3, R2_3]
]

df = pd.DataFrame(model_results, columns=['Model', 'MAE', 'MSE', 'R2'])
print(df)
```

```python
print(np.concatenate((y_pred_rft.reshape(len(y_pred_rft),1),y_test.reshape(len(y_test),1)),1))
```

```python
r2_score(y_train, y_pred_rft_train) # still some overfitting
```

# Model 4(Drop)

```python
from sklearn.linear_model import Ridge
rd=Ridge()
```

```python
parameters = {
    'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]
}

grid_search = GridSearchCV(rd, parameters, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_rd = Ridge(**best_params)
best_rd.fit(X_train, y_train)

y_pred_rd = best_rd.predict(X_test)
```

```python
MAE_4 = mean_absolute_error(y_test, y_pred_rd)
MSE_4 = mean_squared_error(y_test, y_pred_rd)
R2_4 = r2_score(y_test, y_pred_rd)

model_results = [
    ['Logistic Regression', MAE, MSE, R2],
    ['Decision Tree', MAE_1, MSE_1, R2_1],
    ['Random Forest', MAE_3, MSE_3, R2_3],
    ['Ridge Regression', MAE_4, MSE_4, R2_4]
]

df = pd.DataFrame(model_results, columns=['Model', 'MAE', 'MSE', 'R2'])
print(df)

```

# Model 6 (2nd best accuracy)

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr=GradientBoostingRegressor()
```

```python
parameters = {
    'loss': ['squared_error'],
    'learning_rate': [0.1, 0.5, 1],
    'n_estimators': [30, 50, 100, 200],
    'subsample': [0.5, 0.75, 1],
    'criterion': ['friedman_mse', 'mse'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(gbr, parameters, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_gbr = GradientBoostingRegressor(**best_params)
best_gbr.fit(X_train, y_train)

y_pred_gbr = best_gbr.predict(X_test)
```

```python
MAE_6 = mean_absolute_error(y_test, y_pred_gbr)
MSE_6 = mean_squared_error(y_test, y_pred_gbr)
R2_6 = r2_score(y_test, y_pred_gbr)

model_results = [
    ['Logistic Regression', MAE, MSE, R2],
    ['Decision Tree', MAE_1, MSE_1, R2_1],
    ['Random Forest', MAE_3, MSE_3, R2_3],
    ['Gradient Boosting', MAE_6, MSE_6, R2_6]
]

df = pd.DataFrame(model_results, columns=['Model', 'MAE', 'MSE', 'R2'])
print(df)

```
